Sabemos que:
$$
\hat {F} = {\frac{GmM}{|r^2|}} \hat{r} 
$$

Haciendo un poco de algebra obtenemos que:

$$
\hat {a}m = {\frac{GmM}{|r^2|}} \hat{r} 
$$

$$
\hat {a}m = {\frac{GmM}{|r^2|}} \hat{r} 
$$

De aquí obtenemos 2 fórmulas: 

$$
a_{x} = {\frac{GM}{|r^2|}}x 
$$

$$
{a_{y}} = {\frac{GM}{|r^2|}}y 
$$

Sabemos que:

$$
{a_x} = \frac{d{v_x}}{dt} 
$$  

$$ 
{v_x} = \frac{dx}{dt} 
$$

Entonces: 

$$
{a_x} = \frac{d^2x}{dt^2} 
$$

Al final nos quedarán 2 formulas:

$$
x(t) = {x_0}{cos(kt)}
$$

$$
y(t) = {y_0}{sen(kt)}
$$


Defininendo valores:

```julia
x_0 = 5;
```

```julia
y_0 = 2;
```

```julia
k = 100;
```

```julia
t = collect(1:1.5:100);
```

```julia
X = zeros(t);
Y = zeros(t);
```

```julia
for i in eachindex(t)
    X[i] = x_0*cos(k*t[i])
    Y[i] = y_0*sin(k*t[i])
end
```

```julia
using Plots
```

```julia
pyplot()
```

```julia
plot(X,Y,w=2)
title!("Leyes de Kepler")
xaxis!("Posición en X")
yaxis!("Posición en Y")
```

```julia
plot(t,X,w=2)
title!("Leyes de Kepler")
xaxis!("Tiempo")
yaxis!("Posición en X")
```

```julia
plot(t,Y,w=2)
title!("Leyes de Kepler")
xaxis!("Tiempo")
yaxis!("Posición en Y")
```

```julia
variables=hcat(X,Y,t)
```

```julia
writedlm("Leyes-de-Kepler.csv", variables, '|')
```
