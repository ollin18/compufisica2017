```julia
function SMatriz(A,B) 
    if size(A)==size(B)==(2,2) 
        R=zeros(size(A)) #variable para el resultado que debe ser del mismo tamaño que A
        for i in 1:size(A)[1] #por cada elemento del renglon 1 de A
            for j in 1:size(A)[2] #por cada elemento del renglon 2 de A
                c=A[i,j]+B[i,j] #por cada i y j esimo elemento de A en B
                R[i,j]=c #Se colocan los resultados en cada renglon y columna 
            end
        end
        print(R) #Aquí le pedimos que imprima R en la pantalla
    end
end
```

```julia
A=[1 4;6 8;]
```

```julia
b=[4 5;-1 0;]
```

```julia
SMatriz(A,b)
```

```julia
function SMatrizn(A,B) 
    if size(A)==size(B) 
        R=zeros(size(A)) #variable para el resultado debe ser del mismo tamaño que A
        for i in 1:size(A)[1] #por cada elemento del renglon 1 de A
            for j in 1:size(A)[2] #por cada elemento del renglon 2 de A
                c=A[i,j]+B[i,j] #por cada i y j esimo elemento de A en B
                R[i,j]=c #Se colocan los resultados en cada renglon y columna
            end
        end
        print(R) #imprima R en la pantalla
    end
end
```

```julia
g=[8 4 23 1;4 8 1 1;7 1 6 5]
```

```julia
h=[8 7 4  1;8  4 25 6;7 5 9 3]
```

```julia
SMatrizn(g,h)
```
