Una onda es una peturbación que se propaga.Las ondas luminosas no necesitan un medio para propagarse.
la ecuación de una onda luminosa nos informa de la posición (Y) de cualquier punto (X) en cualquier instante (t), por tanto la ecuación de una onda es función de x y de t.

Consideremos una onda originada en un punto (O)y que viaja hacia la derecha con una velocidad consatante (C).Sea (p) un punto cualquiera situado a una distancia X del origen de la onda.Este punto reliza el mismo movimiento que ha relizado el punto (O) despues de un cierto timpo (t_o), es decir, el punto (p) tiene un cierto retraso, que es el tiempo tarda en recorrer la onda la distancia (x).

Como la onda viaja a velocidad constante (c), entoces 

$$
\textbf{c} = \frac{x}{t_o} >>>> \textbf{t_o} = \frac{x}{c}
$$
donde t_o es el retraso entre el punto (p) y el foco emisor (O)

La ecuación de movimiento del foco emisor, es 
$$
 \textbf{Y(v,t)} = Asen\left(\omega(t) + \delta\right)
$$
el punto (p), situado a una distancia (x) del foco, comenzara a moverse más tarde y su estado de vibración será
$$
 \textbf{Y(v,t´)} = Asen\left(\omega(t´) + \delta\right)
$$

tenemos que
$$
\textbf{t´} = t-t_o=t -  \frac{x}{c}
$$
utituyendo en la ecuación anterior, tenemos
$$
 \textbf{Y(v,t´)} = Asen\left(\omega(t -  \frac{x}{c}) + \delta\right)>>>>>> 
$$

$$
 \textbf{Y(v,t´)} = Asen\left((\frac{t2\pi}{T}- \frac{x2\pi}{Tc}) + \delta\right)
$$
puesto que 
$$
\omega =\frac{2\pi}{T} 
$$
entonces 

$$
  \textbf{Y(v,t´)} = Asen\left((\omega(t´)- \frac{x2\pi(f)}{c}) + \delta\right)
$$

$$
\textbf{Y(v,t´)} = Asen\left((\omega(t´)- \frac{x2\pi}{c\lambda}) + \delta\right)
$$
ahora, si hacemos 
$$
\textbf{K} =\frac{2\pi}{\lambda}
$$
entoces,
$$
\textbf{Y(v,t´)} = Asen\left((\omega(t´)- Kx)+ \delta\right)
$$
Esta es la ecuación una onda luminosa en función de t y x,

 donde K es le número de onda. 


```julia
# Grafiquemos primero la onda electromagnética correspondiente al color verde 
# El color verde le corresponde una logitud de onda de 460 nm
δ = π/2 #ésta es la fase inicial
A =10 # A es la aplitud de la onda 
λ =46 #supondremos que 46 son 460 nm
c=300 #la velocidad de las ondas de luz es iagual a 3000000 km/s, supondremos que 300 corresponde a la velocidad de 3000000 km/s
T=λ/c # T es le periodo 
ω =2*π/T #ω es la frecuencia angular 
```

definamos el tiempo como sigue 

```julia
t = collect(0:0.003:2);
```

```julia
X= zeros(t);# nuestra posición en X e Y tienen el mismo número de elementos que el tiempo
Y =zeros(t);
```

```julia
#ahora actualizemos los valores de X e Y en función del tiempo
for i in eachindex(t)
    X[i]=300*t[i]
    Y[i]=A*sin(ω*t[i]+δ)
end
```

```julia
Pkg.add("Plots")
Pkg.add("PyPlot")
```

```julia
using Plots
```

```julia
pyplot()
```

```julia
plot(X,Y,w=4,color=:blue) 
title!("Onda de luz(color azul)")
xaxis!("Posición en X")
yaxis!("Posición en Y")
```

```julia
# Ahora grafiquemos la onda electromagnética correspondiente al color rojo que le corresponde una longitud de onda de 700 nm  
ϕ = 3*π/2 #ésta es la fase inicial
A =10 # A es la aplitud de la onda 
λ_1 =70 #supondremos que 70 son 700 nm
c=300 #la velocidad de las ondas de luz es iagual a 3000000 km/s, supondremos que 300 corresponde a la velocidad de 3000000 km/s
T_1=λ_1/c # T_1 es el periodo 
ω_1 =2*π/T_1 #ω_1 es la frecuencia angular 
```

```julia
S=zeros(t);# usaremos el mismo tiempo definido anteriormente 
Z=zeros(t);
```

```julia
#actualicemos la posion S y Z de la onda en función del tiempo
#ahora actualizemos los valores de X e Y en función del tiempo
for i in eachindex(t)
    S[i]=300*t[i]
    Z[i]=A*sin(ω_1*t[i]+ϕ)
end
```

```julia
plot(S,Z,w=4,color=:red) 
title!("Onda de luz(color rojo)")
xaxis!("Posición en X")
yaxis!("Posición en Y")
```

```julia
 #Si sumamos estas dos ondas, obtenemos 
I=zeros(t);
J=zeros(t);
```

```julia
I=X+S;
J=Y+Z;
```

```julia
plot(I,J,w=4,color=:purple) 
title!("Suma de las ondas de color verde y rojo")
xaxis!("Posición en X")
yaxis!("Posición en Y")
```

## En esta gráfica observamos que los picos más altos corresponden a la interferencia contructiva de las ondas de color azul y rojo y, los picos mas pequeños, en donde existe una interferencia destrutiva.
