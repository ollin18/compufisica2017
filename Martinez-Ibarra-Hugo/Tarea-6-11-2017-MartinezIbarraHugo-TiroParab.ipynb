$$
\mathbf{v_{0}} = v_{0x}\hat{i} + v_{0y}\hat{j} = v_0 cos(\theta_0)\hat{i} + v_0 sen(\theta_0)\hat{j}\qquad(1)
$$
$$
\textbf{a} = a_x\hat{i} + a_y\hat{j} = 0\hat{i} + (-g)\hat{j}\qquad(2)
$$
$$
\textbf{r} = r_x\hat{i} + r_y\hat{j} = (v_{0x}t+x_0)\hat{i} + \left(-\frac{1}{2}gt^2+v_{0y}t+y_0\right)\hat{j}\qquad(3)
$$
$$
\mathbf{v} = v_x\hat{i} + v_y\hat{j} = v_{0x}\hat{i} + (v_{0y}-gt)\hat{j} = v_0 cos(\theta_0)\hat{i} + (v_0 sen(\theta_0)-gt)\hat{j}\qquad(4)
$$
Donde:
$$\mathbf{v_0}:\text{Vector velocidad inicial.}\\
\theta_0:\text{Ángulo de disparo inicial.}\\
\mathbf{a}:\text{Vector aceleración}\\
\mathbf{r}:\text{Vector posición.}\\
\mathbf{v}:\text{Vector velocidad.}
$$
### Tiempo de vuelo

Cuando $$v_y=0$$ se obtiene de (3) y (4):
$$
t = \sqrt{\dfrac{2(r_y-y_0)}{-g}}\qquad(5)
$$
$$
t = \dfrac{v_0 sen(\theta_0)}{g}\qquad(6)
$$
respectivamente, pero para este caso se usará (6); sin emabrgo, ese sólo es el tiempo cuando ha alcanzado su altura máxima, y si se desea tener en cuenta todo el tiempo que está en el aire, dicho tiempo se duplica (debido al commportamiento, de parábola, de la posición del cuerpo), por lo que finalmente se tiene:
$$
t_v = \dfrac{2v_0 sen(\theta_0)}{g}\qquad(7)
$$

```julia
g=9.81 #m/s²
v0=100 #m/s
a=-g;
```

```julia
θ0=collect(0.0:0.0625:π/2);
```

```julia
tv=zeros(θ0);
```

```julia
for i in eachindex(θ0)
    tv[i]=((2.0)*v0*sin(θ0[i]))/g
end
```

```julia
using Plots
```

```julia
gr()
```

```julia
plot(θ0,tv,w=2,label=["tv"],lcolor=:darkred,style=:dashdot)
title!("Tiempo de vuelo")
xaxis!("Ángulo de lanzamiento (rad)")
yaxis!("Tiempo en el aire (s)")
```

### Alcance máximo
De (3) para $r_x$ y $r_y$ con $x_0=0$ y $y_0=0$, respectivamente, se tiene:
$$
r_x=R=v_{0x}t=v_0 cos(\theta_0)t\qquad(8)
$$
Donde R es el alcance máximo.
$$
r_y=0=-\frac{1}{2}gt^2+v_{0y}t=-\frac{1}{2}gt^2+v_0sin(\theta_0)t\qquad(9)
$$
Al eliminar t en ambas ecuaciones (8) y (9) se obtiene:
$$
R=\dfrac{v_0^2}{g}2sin(\theta_0)cos(\theta_0)\qquad(10)
$$
Pero $2sin(\theta_0)cos(\theta_0)=sin(2\theta_0)$, por lo que finalmente se tiene:
$$
R=\dfrac{v_0^2}{g}sin(2\theta_0)\qquad(11)
$$

```julia
R=zeros(θ0);
```

```julia
for i in eachindex(θ0)
    R[i]=((v0^2)/g)*sin(2*θ0[i])
end
```

```julia
gr()
```

```julia
plot(θ0,R,w=2,label=["R"],lcolor=:darkorange,style=:dashdot)
title!("Alcance máximo")
xaxis!("Ángulo de lanzamiento (rad)")
yaxis!("Distancia máxima (m)")
```

```julia
@manipulate for v0=0.0:1.0:100.0
    R=((v0^2)/g)*sin.(2*θ0)
    plot(θ0,R,w=2,xlim=(minimum(θ0),maximum(θ0)),ylim=(minimum(R),maximum(R)),label=["R"])
    title!("Alcance máximo")
    xaxis!("Ángulo de lanzamiento (rad)")
    yaxis!("Distancia máxima (m)")
end
```
