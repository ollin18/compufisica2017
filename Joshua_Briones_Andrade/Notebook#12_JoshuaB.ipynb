**a)** Escribe un programa que arroje como resultado un archivo con los datos de un movimiento rectilíneo uniforme, en un intervalo de tiempo $[t_i,t_f]$ definido por el usuario.

_____________________

La ecuación que tenemos del MRU es la siguiente:

$$ \vec r (t)=\vec r_0 + \vec v t $$

Donde:

$\vec r_0$: es la posición en el tiempo cero con repecto al origen (x,y)

$\vec v$: es la velocidad que es constante

$t$: es el tiempo


```julia
#Definimos nuestra funcion del MRU con las variables de la velocidad, la posicion inicial y el rango de tiempo.
#También nos crea un archivo.csv que se guarda en el directorio donde estemos trabjando.

function MRU(r,v,t)
    R=zeros(t)
    for i in eachindex(t)
        R[i]=(v*t[i])+r
        end
    tabla=hcat(t,R)
    writedlm("MRU.csv", tabla, '|')
    readdlm("MRU.csv", '|')
end
```

```julia
T=MRU(10,3,0:Δt:50)
```

```julia
using Plots;
gr()
```

```julia
plot(T[1:end,2],T[1:end,1],w=5,label=["r(t)"])
title!("MRU")
xaxis!("Tiempo")
yaxis!("Posición respecto al origen")
```

**b)** Escribe un programa que arroje como resultado un archivo con los datos de un movimiento uniformemente acelerado (o caída libre), en un intervalo de tiempo $[t_i,t_f]$ definido por el usuario.

*_______________________*


Continuamos con la gráfica de tiro vertical :)
$$ \vec{r}(t)=\vec{r_0} + \vec{v}t + \frac{\vec{a}t^2}{2} $$

Para analizar con respecto a X y a Y lo separamos en sus componentes.

$$ x(t)=x_0 + v_x t + \frac{a_x t^2}{2} $$

$$ y(t)=y_0 + v_y t + \frac{a_y t^2}{2} $$

Como en el tiro vertical solo se maneja la coordenada Y, despreciaremos a X. También la aceleración es una componente particular porque $a_y =-g=- 9.81 \frac{m}{s^2}$ Entonces nuestro modelo seria el siguiente.
$$ y(t)=y_0 + v_y t - \frac{gt^2}{2} $$

Donde:

$y_0$: es la posición en el tiempo cero con repecto a Y

$v_y$: es la velocidad inicial

$t$: es el tiempo

$g$: es la gravedad


```julia
#Definimos nuestra funcion del tiro vertical con las variables de la velocidad inicial y el rango de tiempo.
#También nos crea un archivo.csv que se guarda en el directorio donde estemos trabjando.
#Suponemos que nuestro Tiro Vertical parte del origen asi que X_0=0 y Y_0=0

function TiroVertical(viy,t)
    Y=zeros(t)
    for i in eachindex(t)
        Y[i]=(viy*t[i])-(9.81*t[i]*t[i])
        end
    tabla=hcat(t,Y)
    writedlm("Tiro_Vertical.csv", tabla, '|')
    readdlm("Tiro_Vertical.csv", '|')
end
```

```julia
T=TiroVertical(300,1:Δt:50)
```

```julia
gr()
```

```julia
plot(T[1:end,1],T[1:end,2],w=2)
title!("Tiro Vertical")
xaxis!("Tiempo (s)")
yaxis!("Posición en Y (m)")
```

_______________

Según el notebook de Ollin, tenemos la siguiente ecuación que rige el moviento del tiro parabólico.
$$ \textbf{r(t)}=(X_0 + V_{0x} t)\hat{i}+ (Y_0 + V_{0y} t - \frac{g t^2}{2})\hat{j}\qquad(1) $$
$$ \textbf{V(t)}=V_x + V_y = v_0cos(\theta)\hat{i}+v_0sin(\theta)\hat{j}\qquad(2) $$

Donde:

$ V_x , V_y $: son las respectivas componentes de la velocidad.

$ \theta $: es el angulo de tiro.

$ \hat{i}, \hat{j} $: son los vectores unitarios de X y Y respectivamente.

```julia
#Definimos nuestra funcion del tiro parabolico con las variables del angulo, la velocidad inicial y el rango de tiempo.
#También nos crea un archivo.csv que se guarda en el directorio donde estemos trabjando.
#Suponemos que nuestro Tiro Parabólico parte del origen asi que X_0=0 y Y_0=0

function TiroParabolico(θ,vi,t)
    X=zeros(t)
    Y=zeros(t)
    vix=vi*cos(θ) #Esto lo definimos asi por la ecuación (2)
    viy=vi*sin(θ)
    for i in eachindex(t)
        X[i]=(vix*t[i])
        Y[i]=(viy*t[i])-(9.81*t1[i]*t1[i])/2  #g=9.81 m/s^2
    end
    tabla=hcat(X,Y,t)
    writedlm("Tiro_Parabólico.csv", tabla, '|')
    readdlm("Tiro_Parabólico.csv", '|')
end
```
