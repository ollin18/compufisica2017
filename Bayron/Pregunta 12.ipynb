```julia
function MRU(t1,t2,Δt,v) #t1 y t2 son el tiempo inicial y final, Δt es la separación entre cada tiempo,v es la velocidad
    t= collect(t1:Δt:t2) #aquí creamos un arreglo de tiemo desde t1 hasta t2 con espacio de Δt entre cada tiempo. 
    posición=zeros(collect(t1:Δt:t2)) #Se crea un arreglo para la posicion
    velocidad=zeros(length(t)) #este para la velocidad
    for i ∈ 1:length(t) #y para cada elemento de t se aplique esto
        if i != 1 #le pedimos que el elemento de t sea distinto de 1 
            posición[i] = posición[i-1]+v*Δt #Aquí utilizamos la ecuación del mru, se le resta 1a i porque manejamos el Δt en la ecuación
        end
    end
    velocidad=velocidad .+ v #Se sumaran todos los elementos de la verlocidad
    mrudatos= hcat(t,posición,velocidad)#Cocatenamos
    writedlm("mrudatos.csv",mrudatos,'|') #Creamos un archivo
    print(t,posición,velocidad) #Y aqui que imprima
end   
```

```julia
function MUA(t1,t2,Δt,v_0,a) #t1 y t2 son los tiempos inicial y final, Δt es el espacio entre cada tiempo, v_0 se refiere a la velocidad inicial y a es la aceleración. 
    t= collect(t1:Δt:t2) #aquí creamos un arreglo de tiemo desde t1 hasta t2 con espacio de Δt entre cada tiempo. 
    posición=zeros(collect(t1:Δt:t2)) #creamos un arreglo de datos de la posición
    velocidad=zeros(collect(t1:Δt:t2)) #creamos un arreglo para la velocidad
    for i ∈ 1:length(t) #A cada elemento de t
        if i != 1 #con t  distinto de 1
            posición[i]=(v_0*t[i])+(1/2*a*(t[i])^2) #utilizamos la ecuación del mua "X(t)=v_0t + 1/2at^2"
            velocidad[i]=v_0+a*t[i] #Aquí utilizamos la ecuación de la velocidad en el mua "V(t)=v_0 + at"
        end
    end
    muac= hcat(t,posición,velocidad)#Aquí concatenamos los datos obtenidos
    writedlm("muac.csv",muac,'|') #Aquí escribimos los datos en un archivo
    print(t,posición,velocidad) #Aquí le pedimos que imprima los datos del tiempo, la posición y la velocidad
end  
```

```julia
function TPara(t1,t2,Δt,θ,g,v_0) #t1,t2 y Δt corresponden al tiempo inicial,final y separación entre cada tiempo,Θ es el angulo, g es el valor de la gravedad, v_0 es la velocidad inicial
    t=collect(t1:Δt:t2) #aquí creamos un arreglo de tiemo desde t1 hasta t2 con espacio de Δt entre cada tiempo.
    X=zeros(t) #arreglo para la posición en X 
    Y=zeros(t) #posición en Y
    for i in eachindex(t) #A todo elemento de T se haga
        X[i] = v_0*cos(θ)*t[i] #Aquí utilizamos la ecuación
        Y[i] = (-(1/2*g*t[i]^2)+v_0*sin(θ)*t[i]) #Aquí utilizamos la ecuación
    end
    TiroPara=hcat(t,X,Y) #concatenamos
    writedlm("TiroPara.csv",TiroPara,'|') #Creamos un archivo
    print(t,X,Y) #Pedimos que imprima
end 
```
