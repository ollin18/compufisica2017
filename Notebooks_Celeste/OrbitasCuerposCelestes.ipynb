Tenemos las siguientes ecuaciones para X,Y en función del tiempo:

$$
X[t] = Xo   cos(kt)
$$

$$
Y[t] = Yo   sen(kt)
$$

Donde k es una constante dada por:

$$
\frac{GM}{|r^2|}
$$

Donde G es la constante de gravitación universal, M es la masa del cuerpo más grande y r es la distancia entre los dos cuerpos.
Por lo tanto k adquiere el valor de 9.81 m/s^2 para el caso de la Tierra.

Y Xo, Yo son las coordenadas iniciales.

Por lo tanto, lo primero es definir nuestras variables asignándoles valores:

```julia
k = 9.81 #m/s^2
```

```julia
Xo = 1
```

```julia
Yo = 1
```

Vamos a necesitar de un tiempo:

```julia
t = collect(0:5:200);
```

Y vamos a tener una posición X,Y para cada tiempo:

```julia
X = zeros(t);
```

```julia
Y = zeros(t);
```

```julia
for i in eachindex(t)
    X = Xo*cos.(k*t)
    Y = Yo*sin.(k*t)
end
    
```

```julia
X;
```

```julia
Y;
```

Finalmente, vamos a graficar:

```julia
using Plots
```

```julia
pyplot()
```

```julia
gr()
```

**Primero vamos a empezar graficando ambas variables X,Y:**

```julia
plot(X,Y,w=2)
title!("Kepler")
xaxis!("X")
yaxis!("Y")
```

**Ahora vamos a graficar X contra tiempo y Y contra tiempo:**

```julia
plot(t,X,w=2)
title!("Kepler")
xaxis!("Tiempo(s)")
yaxis!("X")
    
```

```julia
plot(t,Y,w=2)
title!("Kepler")
xaxis!("Tiempo(s)")
yaxis!("Y")
```

**Y finalmente vamos a escribir nuestros datos en un archivo csv:**

```julia
variables = hcat(X,Y)
```

```julia
writedlm("kepler.csv", variables, '|')
```
