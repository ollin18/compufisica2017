Lo primero que vamos a hacer es instalar todo lo que necesitamos para graficar después:

```julia
Pkg.add("Gadfly")
```

```julia
using Gadfly
```

Ahora si, tenemos la ecuación que describe el movimiento rectinlíneo uniforme:

**posición[i+1] = posición[i] + velocidad*Δt**

El usuario puede elegir tanto el tiempo como la velocidad con la que quiere trabajar. En este caso elegimos arbitrariamente, pero siempre se le puede asignar un nuevo valor al anterior.

Entonces, definimos primero un intervalo de tiempo, el Δt indica cada cuanto avanza dentro de ese intervalo.

```julia
Δt = 0.125
tiempo = 0:Δt:100 
```

Una velocidad:

```julia
velocidad = 96 #m/s
```

Como la posición va ir cambiando dependiendo del tiempo, empezamos con arreglo de ceros que sea justo del tamaño del tiempo:

```julia
posición = zeros(length(tiempo));
```

Y escribimos nuestro programa, basándonos en la ecuación anterior:

```julia
for i ∈ 1:length(tiempo) #Va ir recorriendo el arreglo del tiempo
    if i != 1 #Nos interesa cuando no es 1, porque en 1, la posición es la misma
        posición[i] = posición[i-1]+velocidad*Δt #m
    end
end
```

Ahora, para comprobar que está bien definido el tiempo:

```julia
tiempo
```

Observamos los valores para la posición que nos ha arrojado el programa:

```julia
posición
```

Hacemos un collect para el tiempo:

```julia
Tiempo = collect(tiempo)
```

Y finalemente, procedemos a graficar nuestros datos de posición y velocidad:

```julia
plot(x=Tiempo, y=posición, Geom.line, Guide.xlabel("Tiempo(s)"), Guide.ylabel("Posición"), Guide.title("MRU"))
```

```julia
velocidad_arreglo = zeros(length(tiempo))
velocidad_arreglo = velocidad_arreglo .+ velocidad
```

```julia
plot(x=Tiempo, y=velocidad_arreglo, Geom.line, Guide.xlabel("Tiempo(s)"), Guide.ylabel("Velocidad"), Guide.title("MRU"))
```
