Ecuación de la recta

$$ 
y=mx+b
$$

Le damos valores a m y a b

```julia
m=10
```

```julia
b=19
```

```julia
x = collect(-20:0.125:20);
```

```julia
y = zeros(x);
```

```julia
for i in eachindex(x)
    y = m*x+b
    
end
```

```julia
using Plots
```

```julia
pyplot()
```

```julia
plot(x,y,w=2)
title!("Recta")
xaxis!("x")
yaxis!("y")
```

```julia
variables = hcat(x,y)
```

```julia
writedlm("graficarecta.csv",variables,'|')
```
