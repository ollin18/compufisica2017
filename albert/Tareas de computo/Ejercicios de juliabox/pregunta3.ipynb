Siguiendo la siguiente serie

$\displaystyle{\sum_{n=0}^{∞} i} = \dfrac{{-1}^n}{2n+1} = \dfrac {π}{4}$

Escribe un programa que calcule diferentes aproximaciones de π para diferentes valores.

```julia
function PI(n)
        pi=0 #se hace un contador 
        co=0 #llega el contador mas 1 hasta llegar a n 
        while n>=co #inicio del bucle 
            pi=pi+((-1)^co)/(2co+1) #Se actualiza el contador hasta llegar al bucle 
            co+=1 #hacemos el contador para poder alcanzar el vucle 
        end 
    print(4pi) #se imprime el resultado 
end
```
