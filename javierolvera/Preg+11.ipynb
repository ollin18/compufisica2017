a) Este programa grafica los puntos de una recta dados los valores de la pendiente, ordenada al origen, así como los valores inicial, final y el tamaño del intervalo para x.

```julia
 using Plots #LLamando las librerías para graficar
    pyplot()
```

```julia
function recta(m,b,x0,xf,Δx) #Genera en una sola funcion que depende de todas las variables del sistema, los requerimientos para la recta 
    x = collect(x0:Δx:xf) #Ubica a x en un intervalo con Δx como incremento
    y = zeros(x)          #Genera un vector y igual a x, pero rellenado de ceros
    for i in eachindex(x)
    y[i] = m*x[i] + b      #Generamos un valor de y para cada x con la ecuación ordinaria de la recta
    end
   
    plot(x,y,w=2)
title!("Gráfica de una recta")
xaxis!("Eje x")
yaxis!("Eje y")
end
```

```julia
recta(1,0,0,1,0.01)
```

b) Este programa grafica los puntos de una parábola dados los valores de los coeficientes del término cuadrático, lineal e independiente, así como los valores inicial, final y el tamaño del intervalo para x.

```julia
function parábola(a,b,c,x0,xf,Δx)
    x = collect(x0:Δx:xf)
    y = zeros(x)
    for i in eachindex(x)
    y[i] = a*x[i]^2 + b*x[i] + c #Usamos la forma de trinomio cuadrado 
    end
variables = hcat(x,y) #Concatenamos horizontalmente ambos vectores para generar la "tabulación"
writedlm("parábola.csv", variables, '|') #Creamos un archivo y escribimos en el la tabla creada
datos = readdlm("parábola.csv", '|') #verificamos la creación del archivo leyendolo y guardando el resultado de la lectura en datos
X = datos[:,1] #Desconcatenamos para asignar la primer columna a X y la segunda a Y
Y = datos[:,2]
plot(X,Y,w=2)  #Mandamos graficar las  nuevas variables
title!("Gráfica de una parábola")
xaxis!("Eje x")
yaxis!("Eje y")
end
```

```julia
parábola(1,2,1,-5,5,0.125)
```
