<font color="#385BBA" size="13"> Peonza (trompo) 

```julia
θ=collect(0:π/32:2π);
```

```julia
x=zeros(θ);
y=zeros(θ);
```

```julia
for i ∈ eachindex(θ)
    x=cos(θ)
    y=sin(θ)
end
```

```julia
using Plots
```

```julia
plot(x,y,w=2)
title!("Circunferencia que marca el paso del trompo")
```

    Que en el momento de poder tener un radio mas grande nos esta indicando que la poscicion del centro de masas esta cambiando, o de otra manera el angulo generado por la vertical dimencional que en este caso seria el centro del circulo, se esta haciendo mas grande o mas pequeño.

```julia
p=collect(0:2:100);
```

```julia
α=π/4
```

```julia
m=tan(α)
```

```julia
b=1
```

```julia
y=zeros(p);
```

```julia
for i ∈ eachindex(p)
    y=m*p+b
end
```

```julia
using Plots
```

```julia
plot(p,y,w=2)
title!("Relacion entre la masa y el tiempo maximo de rotacion")
xaxis!("Tiempo")
yaxis!("Masa")
```
