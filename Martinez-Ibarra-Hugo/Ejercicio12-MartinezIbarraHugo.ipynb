a) Escribe un programa que arroje como resultado un archivo con los datos de un movimiento rectilíneo uniforme, en un intervalo de tiempo [ti,tf] definido por el usuario.

$$x = vt+x_0$$

```julia
#Primero se llama al paquete para graficar Plots: GR.
using Plots
gr()
```

```julia
#Se define el tamaño de paso del rango de valores del tiempo, luego el valor inicial y final del rango de valores de t. Después se define dicho rango de valores de tiempo con la sintaxis adecuada. Finalmente se da un valor de la velocidad.
Δt=1 #s
ti=0 #s
tf=100 #s
t=ti:Δt:tf #s
v=10; #m/s
```

```julia
#Se crea un vector de ceros que alamcenará los datos de posición (x), que tiene el mismo tamaño de entradas que el rango de valores de tiempo.
#Con un ciclo for se llena cada entrada del vector x, al indicar que el primer dato de posición es cero, porque el primer valor de tiempo es cero, para los valores siguientes se define el valor de posición siguiente como el valor de posición anterior más la velocidad por la diferencia de tiempo entre estas dos posiciones.
x=zeros(t)
for i ∈ 1:length(t)
    if i != 1
        x[i] = x[i-1]+v*Δt
    end
end
x #m;
```

```julia
#Se escriben los valores de t y x en archivos que luego se leerán.
writedlm("t.txt",t,'|')
writedlm("x.txt",x,'|')
#Se leen los datos provenientes de los archivos.
t=readdlm("t.txt",'|')
x=readdlm("x.txt",'|');
```

```julia
#Con la función plot(), se indica que arreglo (vector en este caso) contiene los valores de las coordendas x y qué arreglo contiene los valores de las coordenadas y, con w=2 se indica el grosor de línea, luego con label la etiqueta del gráfico y con lcolor el color de la línea de gráfico. Al final se añaden los títulos de los ejes.
plot(t,x,w=2,label=["x=f(t)"],lcolor=:red)
title!("MRU")
xaxis!("t [s]")
yaxis!("x [m]")
```

```julia
#Otra forma es que cada dato de posición está definido por el valor incial de posición, que se elige como cero más la velocidad por cada valor de tiempo sucesivo.
x=zeros(t)
for i ∈ 1:length(t)
    x[i] = x[1]+v*t[i]
end
x #m;
```

```julia
#Se escriben los valores de t y x en archivos que luego se leerán.
writedlm("t.txt",t,'|')
writedlm("x.txt",x,'|')
#Se leen los datos provenientes de los archivos.
t=readdlm("t.txt",'|')
x=readdlm("x.txt",'|');
```

```julia
#Con la función plot(), se indica que arreglo (vector en este caso) contiene los valores de las coordendas x y qué arreglo contiene los valores de las coordenadas y, con w=2 se indica el grosor de línea, luego con label la etiqueta del gráfico y con lcolor el color de la línea de gráfico. Al final se añaden los títulos de los ejes.
plot(t,x,w=2,label=["x=f(t)"],lcolor=:red)
title!("MRU")
xaxis!("t [s]")
yaxis!("x [m]")
```

b) Escribe un programa que arroje como resultado un archivo con los datos de un movimiento uniformemente acelerado (o caída libre), en un intervalo de tiempo [ti,tf] definido por el usuario.

$$y_{f}=-\frac{1}{2}gt^2+v_{0}t+y_i$$

```julia
#Se define el tamaño de paso del rango de valores del tiempo, luego el valor inicial y final del rango de valores de t. Después se define dicho rango de valores de tiempo con la sintaxis adecuada. Finalmente se da un valor de la velocidad.
Δt=0.05 #s
ti=0 #s
tf=2.05 #s
t=ti:Δt:tf #s
v0=0 #m/s
g=9.81; #m/s^2
```

```julia
#Primero se crea un vector de ceros para los datos de posición finales (yf), que sea del mismo tamaño que el rango de valores del tiempo. Con un ciclo for que por cada entrada del rango de valores del tiempo se toma el respectivo valor de tiempo, el valor constante de velocidad inicial y el valor inicial de posición (elegido como cero) para obtener un valor de posición final.
yf=zeros(t)
for i ∈ 1:length(t)
    yf[i]=yf[1]+v0*t[i]+0.5*(-g)*(t[i])^2
end
yf;
```

```julia
#Se escriben los valores de t y x en archivos que luego se leerán.
writedlm("t.txt",t,'|')
writedlm("yf.txt",yf,'|')
#Se leen los datos provenientes de los archivos.
t=readdlm("t.txt",'|')
yf=readdlm("yf.txt",'|');
```

```julia
#Con la función plot(), se indica que arreglo (vector en este caso) contiene los valores de las coordendas x y qué arreglo contiene los valores de las coordenadas y, con w=2 se indica el grosor de línea, luego con label la etiqueta del gráfico y con lcolor el color de la línea de gráfico. Al final se añaden los títulos de los ejes.
plot(t,yf,w=2,label=["yf=f(t)"],lcolor=:red)
title!("Caída libre")
xaxis!("t [s]")
yaxis!("yf [m]")
```

c) Escribe un programa que arroje como resultado un archivo con los datos de un tiro parabolico, en un intervalo de tiempo [ti , tf] definido por el usuario.

$$
\mathbf{v_{0}} = v_{0x}\hat{i} + v_{0y}\hat{j} = v_0 cos(\theta_0)\hat{i} + v_0 sen(\theta_0)\hat{j}\qquad(1)
$$
$$
\textbf{a} = a_x\hat{i} + a_y\hat{j} = 0\hat{i} + (-g)\hat{j}\qquad(2)
$$
$$
\textbf{r} = r_x\hat{i} + r_y\hat{j} = (v_{0x}t+x_0)\hat{i} + \left(-\frac{1}{2}gt^2+v_{0y}t+y_0\right)\hat{j} = (v_0 cos(\theta_0)t+x_0)\hat{i} + \left(-\frac{1}{2}gt^2+v_0 sen(\theta_0)t+y_0\right)\hat{j}\qquad(3)
$$
$$
\mathbf{v} = v_x\hat{i} + v_y\hat{j} = v_{0x}\hat{i} + (v_{0y}-gt)\hat{j} = v_0 cos(\theta_0)\hat{i} + (v_0 sen(\theta_0)-gt)\hat{j}\qquad(4)
$$
Donde:
$$\mathbf{v_0}:\text{Vector velocidad inicial.}\\
\theta_0:\text{Ángulo de disparo inicial.}\\
\mathbf{a}:\text{Vector aceleración}\\
\mathbf{r}:\text{Vector posición.}\\
\mathbf{v}:\text{Vector velocidad.}
$$

```julia
#Se define el tamaño de paso del rango de valores del tiempo, luego el valor inicial y final del rango de valores de t. Después se define dicho rango de valores de tiempo con la sintaxis adecuada. Finalmente se da un valor de la velocidad.
Δt=0.05 #s
ti=0 #s
tf=15 #s
t=ti:Δt:tf #s
v0=100 #m/s
g=9.81 #m/s^2
θ=π/4 #rad;
```

```julia
#Primero se crean dos vectores de ceros, cada uno almacenará eventualmente los datos de posición en la coordenada x e y.
#Con un ciclo for se llena cada entrada que posición para la coordenada x e y al tiempo que se toma cada valor correspondiente del rango de valores del tiempo; de acuerdo a las ecuaciones descritas.
x=zeros(t)
y=zeros(t)
for i in eachindex(t)
    x[i]=v0*cos(θ)*t[i]+x[1]
    y[i]=(-(0.5*g*t[i]^2)+v0*sin(θ)*t[i])
end
```

```julia
#Se escriben los valores de t y x en archivos que luego se leerán.
writedlm("x.txt",x,'|')
writedlm("y.txt",y,'|')
#Se leen los datos provenientes de los archivos.
x=readdlm("x.txt",'|')
y=readdlm("y.txt",'|');
```

```julia
#Con la función plot(), se indica que arreglo (vector en este caso) contiene los valores de las coordendas x y qué arreglo contiene los valores de las coordenadas y, con w=2 se indica el grosor de línea, luego con label la etiqueta del gráfico y con lcolor el color de la línea de gráfico. Al final se añaden los títulos de los ejes.
plot(x,y,w=2,label=["y=f(x)"],lcolor=:red)
title!("Tiro parabólico")
xaxis!("x [m]")
yaxis!("y [m]")
```

### Tiempo de vuelo

Cuando $$v_y=0$$ se obtiene de (3) y (4):
$$
t = \sqrt{\dfrac{2(r_y-y_0)}{-g}}\qquad(5)
$$
$$
t = \dfrac{v_0 sen(\theta_0)}{g}\qquad(6)
$$
respectivamente, pero para este caso se usará (6); sin emabrgo, ese sólo es el tiempo cuando ha alcanzado su altura máxima, y si se desea tener en cuenta todo el tiempo que está en el aire, dicho tiempo se duplica (debido al commportamiento, de parábola, de la posición del cuerpo), por lo que finalmente se tiene:
$$
t_v = \dfrac{2v_0 sen(\theta_0)}{g}\qquad(7)
$$

```julia
#Aquí se va a variar el ángulo inicial de lanzamiento.
g=9.81 #m/s²
v0=100 #m/s
a=-g;
θ0=collect(0.0:0.0625:π/2);
```

```julia
#Se crea un vector que almacenará los datos de tiempo de vuelo y se llena con un ciclo for para cada valor de ángulo.
tv=zeros(θ0)
for i in eachindex(θ0)
    tv[i]=((2.0)*v0*sin(θ0[i]))/g
end
```

```julia
#Se grafica el tiempo de vuelo en función del ángulo de lanzamiento.
plot(θ0,tv,w=2,label=["tv"],lcolor=:darkred,style=:dashdot)
title!("Tiempo de vuelo")
xaxis!("Ángulo de lanzamiento (rad)")
yaxis!("Tiempo en el aire (s)")
```

```julia
#Para usar una paquetería para gráficos con cambio de parámetros en tiempo real.
using Interact
```

```julia
#Se usa la paquetería anterior más el uso del broadcast, un tipo de ciclo para llenar los datos de tiempo de vuelo.
@manipulate for v0=0.0:1.0:100.0
    tv=((2.0)*v0*sin.(θ0))/g
    plot(θ0,tv,w=2,xlim=(minimum(θ0),maximum(θ0)),ylim=(minimum(tv),maximum(tv)),label=["tv"])
    title!("Tiempo de vuelo")
    xaxis!("Ángulo de lanzamiento (rad)")
    yaxis!("Tiempo en el aire (s)")
end
```

```julia
#Aquí se va a varia el valor de la velocidad inicial o de lanzamiento. Y se llena cada valor de tiempo de vuelo con un ciclo for por cada valor de velocidad de lanzamiento.
v_0=collect(0.0:1.0:100.0)
t_v=zeros(v_0)
gr();
θ_0=π/8;
for i in eachindex(v_0)
    t_v[i]=((2.0)*v_0[i]*sin(θ_0))/g
end
```

```julia
#Se grafica el tiempo de vuelo en función de la velocidad de lanzamiento.
plot(v_0,t_v,w=2,label=["tv"],lcolor=:darkred,style=:dashdot)
title!("Tiempo de vuelo")
xaxis!("Velocidad de lanzamiento (m/s)")
yaxis!("Tiempo en el aire (s)")
```

### Alcance máximo
De (3) para $r_x$ y $r_y$ con $x_0=0$ y $y_0=0$, respectivamente, se tiene:
$$
r_x=R=v_{0x}t=v_0 cos(\theta_0)t\qquad(8)
$$
Donde R es el alcance máximo.
$$
r_y=0=-\frac{1}{2}gt^2+v_{0y}t=-\frac{1}{2}gt^2+v_0sin(\theta_0)t\qquad(9)
$$
Al eliminar t en ambas ecuaciones (8) y (9) se obtiene:
$$
R=\dfrac{v_0^2}{g}2sin(\theta_0)cos(\theta_0)\qquad(10)
$$
Pero $2sin(\theta_0)cos(\theta_0)=sin(2\theta_0)$, por lo que finalmente se tiene:
$$
R=\dfrac{v_0^2}{g}sin(2\theta_0)\qquad(11)
$$

```julia
#Aquí se va a ver el alcance máximo debido al cambio de ángulo de lanzamiento.
R=zeros(θ0)
for i in eachindex(θ0)
    R[i]=((v0^2)/g)*sin(2*θ0[i])
end
```

```julia
#Se grafica el alcance máximo en función del ángulo de lanzamiento.
gr()
plot(θ0,R,w=2,label=["R"],lcolor=:darkorange,style=:dashdot)
title!("Alcance máximo")
xaxis!("Ángulo de lanzamiento (rad)")
yaxis!("Distancia máxima (m)")
```

```julia
#Se verá el alcance máximo debido al cambio del ángulo de lanzamiento (inicial).
@manipulate for v0=0.0:1.0:100.0
    R=((v0^2)/g)*sin.(2*θ0)
    plot(θ0,R,w=2,xlim=(minimum(θ0),maximum(θ0)),ylim=(minimum(R),maximum(R)),label=["R"])
    title!("Alcance máximo")
    xaxis!("Ángulo de lanzamiento (rad)")
    yaxis!("Distancia máxima (m)")
end
```

```julia
#Aquí se verá el alcance máximo debido a la variación de la velocidad de lanzamiento.
Rm=zeros(v_0)
gr()
θ_0=π/8;
for i in eachindex(v_0)
    Rm[i]=((v_0[i]^2)/g)*sin(2*θ_0)
end
```

```julia
#Se grafica el alcance máximo en función de la velocidad de lanzamiento (inicial).
plot(v_0,Rm,w=2,label=["R"],lcolor=:darkorange,style=:dashdot)
title!("Alcance máximo")
xaxis!("Velocidad de lanzamiento (m/s)")
yaxis!("Distancia máxima (m)")
```
