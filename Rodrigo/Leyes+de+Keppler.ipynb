```julia
t = collect(0:5.0:500)
```

```julia
k=54
```

```julia
x_0 = 1
```

```julia
y_0 = 1
```

```julia
X=zeros(t);
Y=zeros(t);
```

```julia
t[3]
```

```julia
for i ∈ eachindex(t)
    X[i]=x_0*cos(k*t[i])
    Y[i]=y_0*sin(k*t[i])
end
```

```julia
using Plots
```

```julia
pyplot()
```

```julia
plot(X,Y,w=2)
title!("Leyes de Keppler")
xaxis!("posicion en x (m)")
yaxis!("posicion en y (m)")
```

```julia
plot(X,t,w=2)
title!("Leyes de Keppler")
xaxis!("posicion en x (m)")
yaxis!("posicion en y(t) (m)")
```

```julia
plot(Y,t,w=2)
title!("Leyes de Keppler")
xaxis!("posicion en t(x) (m)")
yaxis!("posicion en y (m)")
```

```julia
variables=hcat(X,Y,t)
```

```julia
writedlm("leyes-de-Keppler.csv" ,variables, '|')
```
