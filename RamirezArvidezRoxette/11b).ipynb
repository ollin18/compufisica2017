Para la parábola, utilizaremos la fórmula en forma canónica. (Considerando quetiene vértice en el origen)
$$ y^2=4px$$

Con:

$y$ = variable dependiente

$x$ = variable independiente

$p$ = distancia de el vértice al foco

Despejando:
$$ y=\pm2\sqrt{px}$$


```julia
x=collect(0:0.125:50); #Definimos el rango de x
```

```julia
p=4
```

```julia
    y1=zeros(x);  #Se considerá y1 y y2, para negativos y positivos. SE generan los arreglos con ceros de dimensionnes de x,ya que y es dependiente de x 
    y2=zeros(x);
```

```julia
for i in eachindex(x) #Para cada elemento que corresponde a y1 y y2 se realiza la operación
        y1[i]=+2*((p*x[i])^(1/2))
        y2[i]=-2*((p*x[i])^(1/2))
end
```

```julia
Variables=hcat(x,y1,y2)
```

```julia
writedlm("Parabola.csv",Variables,'|');
   
```

```julia
readdlm("Parabola.csv",'|');
```
