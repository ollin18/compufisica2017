Según el notebook de Ollin, tenemos la siguiente ecuación que rige el moviento del tiro parabólico.

$$ \textbf{r(t)}=(X_0 + V_{0x} t)\hat{i}+ (Y_0 + V_{0y} t - \frac{g t^2}{2})\hat{j}\qquad(1) $$

$$ \textbf{V(t)}=V_x + V_y = v_0cos(\theta)\hat{i}+v_0sin(\theta)\hat{j}\qquad(2) $$

Donde:

$ V_x , V_y $: son las respectivas componentes de la velocidad.

$ \theta $: es el angulo de tiro.

$ \hat{i}, \hat{j} $: son los vectores unitarios de X y Y respectivamente.

### Tiempo de Vuelo

Para poder analizar el tiempo de vuelo, lo unico que nos importa es la componente de la velocidad en Y por ello calculamos el tiempo cuando $V_y = 0$.  
Entonces suponemos que $\hat{i}=0$ y $Y_0=0$ en la ecuación (1), por tanto.

$$ \textbf{r(t)}=0=V_{0y}t -\frac{g t^2}{2}\qquad(3) $$

Ahora tenemos una ecuación de segundo grado con dos soluciones. La primera es t=0, que corresponde al instante inicial. La que buscamos es la segunda solución:

$$ t=\frac{2V_{0y}}{g}\qquad(4) $$

También si analizamos gráficamente nos damos cuenta que: $sen(\alpha)=\frac{V_{0y}}{V_0}$

Así que al despejar $V_{0y}$ tenemos: $$V_{0y}=V_0sen(\alpha)\qquad(5)$$

Así que por (4) y por (5) tenemos que el tiempo de vuelo es:
$$ t=\frac{2V_0sen(\alpha)}{g} $$

### Continuamos con las gráficas

```julia
#Primero tomaremos al angulo como la variable asi que definimos las constantes a usar.

g=9.81 #m/s^2
V_0=100 #m/s
```

```julia
#Como el tiempo de vuelo esta en función del angulo de tiro hacemos un arreglo del angulo 
#y declaramos el tiempo de vuelo como un arreglo de las mismas dimensiones.

Angulo=collect(0:0.0125:π/2)
TiempoVuelo=zeros(Angulo);
```

```julia
#Hacemos nuestro codigo según la formula de tiempo de vuelo.

for i in eachindex(Angulo)
    TiempoVuelo[i]=(2*V_0*sin(Angulo[i]))/g
end
```

```julia
using Plots
```

```julia
gr()
```

```julia
plot(Angulo,TiempoVuelo,w=2,label=["T. Vuelo"])
title!("Tiempo de Vuelo respecto al Angulo")
xaxis!("Angulo de Tiro")
yaxis!("Tiempo en caer")
```

### Alcance Máximo

Ahora vamos con la altura máxima

Tomamos la ecuación de la posición de nuestro tiro parabólico, pero solo nos enfocaremos en el eje X, y suponemos que parte de 0.

$$ \textbf{r(t)}= X(t) =V_{0x} t $$

Ahora nos enfocamos en sustituir el tiempo de la ecuación por el tiempo de vuelo y tenemos el alcance máximo.

$$ X_{max}=\frac{2V_0^2sen(\alpha)cos(\alpha)}{g} $$

```julia
#Primero tomamos al angulo como variable

g=9.81 #m/s^2
V_0=100.0 #m
```

```julia
#declaramos al angulo como un arreglo, y Xmax tiene las mismas dimensiones.

Angulo=collect(0:π/32:π/2)
Xmax=zeros(Angulo);
```

```julia
for i in eachindex(Angulo)   
    Xmax[i]=(2*V_0^2*sin(Angulo[i]).*cos(Angulo)[i])/g
end
```

```julia
gr();
plot(Angulo,Xmax,w=2,label=["Tiro parabólico respecto a la vel."])
title!("Tiro Parbólico desplazamiento")
xaxis!("Angulo de Tiro")
yaxis!("Desplazamiento máximo")
```

```julia
#Ahora tomamos a un angulo fijo

g=9.81 #m/s^2
Angulo=π/4
```

```julia
#declaramos a la velocidad como un arreglo y Xmax tiene las mismas dimensiones.

V_0=collect(0.0:100.0);
Xmax=zeros(V_0);
```

```julia
for i in eachindex(V_0)
    Xmax[i]=(2.0*(V_0[i])^2*sin(Angulo)*cos(Angulo))/g
end
```

```julia
gr();
plot(V_0,Xmax,w=2,label=["Alcance máximo"])
title!("Tiro Parabólico pi/2")
xaxis!("Velocidad inicial")
yaxis!("Distacia desplazada")
```

```julia
@manipulate for Angulo=0:π/64:π/2
    for i in eachindex(V_0)
    Xmax[i]=(2.0*(V_0[i])^2*sin(Angulo)*cos(Angulo))/g
    end
    plot(V_0,Xmax,w=2,label=["Alcance máximo"])
    title!("Tiro Parabólico pi/2")
    xaxis!("Velocidad inicial")
    yaxis!("Distacia desplazada")
end
```
