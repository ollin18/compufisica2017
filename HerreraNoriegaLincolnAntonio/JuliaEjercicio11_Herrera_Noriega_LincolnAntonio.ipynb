a) Calcular los valores de una recta para ciertos rangos de las dos variables, y entregar un archivo con los datos a graficar

```julia
using Plots
pyplot()
```

Definimos la función 

```julia
function recta(m,b,x0,xf,Δx)
    x = collect(x0:Δx:xf)     # array del tamaño e intervalos establecidos por el usuario
    y = zeros(x)              # array de ceros del tamaño de x
    for i in eachindex(x)     # el contador determinado por el intervalo x
        y[i] = m*x[i] + b     # obtenemos los puntos de la recta con un ciclo for de la expresión sobre el contador establecido
    end
    variables = hcat(x,y)                     # concatenamos las variables para poder escribirlas
    writedlm("recta.csv", variables, '|')     # escribimos variables con | como separador en el archivo recta.csv
end
```

Y ejecutamos para crear el archivo de datos

```julia
recta(0.5,0,0,1,0.01);
```

b) Calcular los valores de la parábola y entregar un archivo con los datos a graficar

```julia
function parábola(a,b,c,x0,xf,Δx)
    x = collect(x0:Δx:xf)
    y = zeros(x)
    for i in eachindex(x)
        y[i] = a*x[i]^2 + b*x[i] + c
    end
    variables = hcat(x,y)
    writedlm("parábola.csv", variables, '|')
end
```

Y ejecutamos para crear el archivo de datos

```julia
parábola(1,1,1,-2,1,0.01)
```

Graficamos la recta leyendo el archivos de datos creado

```julia
a=readdlm("recta.csv",'|')       # leemos el archivo recta.csv interpretando | como separador
X = a[:,1]                       # El vector X como todos los renglones en la primer columna
Y = a[:,2]                       # El vector y como todos los renglones en la segunda columna
plot(X,Y,w=2)
title!("Gráfica de una recta")
xaxis!("Eje x")
yaxis!("Eje y")
```

Graficamos la parábola leyendo el archivo de datos

```julia
a=readdlm("parábola.csv",'|')
X = a[:,1]
Y = a[:,2]
plot(X,Y,w=2)
title!("Gráfica de una parábola")
xaxis!("Eje x")
yaxis!("Eje y")
```
