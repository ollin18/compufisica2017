```julia
t=collect(0:3:200);
```

```julia
k=10
```

```julia
x_0=1
```

```julia
y_0=1
```

```julia
x=zero(t);
y=zero(t);
```

```julia
for i âˆˆ eachindex(t)
    x=x_0*cos(k*t)
    y=y_0*sin(k*t)
end
```
