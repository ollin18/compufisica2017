__Potencial eléctrico de un dipolo:__

## $$\phi(r,\theta)=\frac{kp}{r^2}\bigg (cos\theta \vec r \bigg)$$

```julia
#Se usa la paquetería para graficar de Gadfly y se desactiva el modo interactivo para poder graficar sin problema.
using PyPlot
ioff()
```

```julia
# Definimos los valores necesarios para trabajar con nuestro dipolo
l=0.25 # m La semidistancia que separa las cargas opuestas.
q=1.9e-12 # C La magnitud de cada carga. Recordar que para un dipolo, son de la misma magnitud pero "signo opuesto" (éste ya está considerado en la ecuación de potencial)
k=9.0e9 # Nm²/C² 
p1=2*l*q # mC
k*q # Nm²/C El producto de kq es sólo para dar la idea de la magnitud del valor del potencial.
θ = [0:2pi/60:2pi;]; #Angulos y el intervalo con el que se van a trabajar
```

```julia
# Definimos valores de radio r en su respectivo vector y definimos los arreglos de lineas equipotenciales con los que se va a trabajar, tenemos agregado una constante k^3 como divisor para hacer que se suavice la gráfica y podamos apreciarla mejor.
r=collect(1:15)
ϕ1 = zeros(length(θ),length(r))
ϕ2 = zeros(length(θ),length(r))
for i in eachindex(θ)
    for j in 1:size(ϕ1)[2]
        ϕ1[i,j]=+(2*cos(θ[i]))*(k*p1/k^3*(r[j])^2)
        ϕ2[i,j]=-(2*cos(θ[i]))*(k*p1/k^3*(r[j])^2)
    end
end
ϕ=hcat(θ,ϕ1,ϕ2);
```

```julia
fig = figure("pyplot_windrose_lineplot",figsize=(10,10)) # Crea una nueva figura
ax = axes(polar="true") # Configura el arreglo en coordenadas polares
title("Lineas equipotenciales ϕ")
p = plot(ϕ[1:end,1],ϕ[1:end,2:end],linestyle="-",marker="None") # Entrada basica para plotear en PyPlot

dtheta = 10
ax[:set_thetagrids]([0:dtheta:360-dtheta;]) # Enseña lineas divisoras en cada incremento dtheta de 0 hasta 360
ax[:set_theta_zero_location]("N") # Definimos 0 grados o "North" en la cima de la gráfica
ax[:set_theta_direction](-1) # Que avance en sentido horario
```

__Comparación:__ Las lineas rojas representan el potencial eléctrico mientras que las lineas azules, el campo eléctrico.

__Campo eléctrico de un dipolo:__

## $$E_e(r_0,\theta)=r_0 sin^2 \theta $$

```julia
# Definimos valores de radio r_0 (es decir el radio cuando θ = 0 grados de la linea de campo) en su respectivo vector y definimos los arreglos de lineas de campo con los que se va a trabajar.
r0=collect(1:10)
E1 = zeros(length(θ),length(r0))
for i in eachindex(θ)
    m=1
    for j in 1:size(E1)[2]
        E1[i,j]=+(sin(θ[i]))*(sin(θ[i]))*r0[j]
        m=m+1
    end
end
E=hcat(θ,E1)
```

```julia
fig = figure("pyplot_windrose_lineplot",figsize=(10,10)) # Crea una nueva figura
ax = axes(polar="true") # Configura el arreglo en coordenadas polares
title("Lineas de campo eléctrico")
p = plot(E[1:end,1],E[1:end,2:end],linestyle="--",marker="None") # Entrada basica para plotear en PyPlot

dtheta = 10
ax[:set_thetagrids]([0:dtheta:360-dtheta;]) # Enseña lineas divisoras en cada incremento dtheta de 0 hasta 360
ax[:set_theta_zero_location]("N") # Definimos 0 grados o "North" en la cima de la gráfica
ax[:set_theta_direction](-1) # Que avance en sentido horario
```

__Comparación:__ Las lineas rojas representan el potencial eléctrico mientras que las lineas azules, el campo eléctrico.
