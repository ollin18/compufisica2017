Sabemos que la ecuación de la recta es: $y=mx+b$ entonces crearemos una función dónde ingreses $m$, $b$ y el intérvalo de $x$

Antes correremos los paquetes para graficar.

```julia
using Plots;
pyplot();
gr();
```

Ahora hagamos la función:

```julia
function Recta(xi,x1,x2,b,m) #ingresamos las variables, dónde x1-x2 es el intervalo, xi cómo será dividido este intervalo  
    x=collect(x1:xi:x2) #Este es el intervalo
    println("Los valores de y en el intérvalo de $x1 a $x2 son:")
    y=zeros(size(x)) #Creamos un arreglo para que vaya guardando cada valor de la ecuación
    for i ∈ eachindex(x)
        y[i]=(m*x[i]+b) #Aquí tenemos la ecuación en función del tiempo
        println(y[i]) 
    end
    plot(x,y,w=1) #La gráfica
        title!("Línea Recta")
        xaxis!("Eje x")
        yaxis!("Eje y")
end        
```

```julia
Recta(1.0,1,20.0,5,0.25)
```

La ecuación de la parábola es: $ {A{x^2}}+Bx+C $

```julia
function Parabola(A,B,C,xi,x1,x2) #Las x funcionan de la misma manera del codigo de arriba
    x=collect(x1:xi:x2)
    println("Los valores de y en el intérvalo de $x1 a $x2 son:")
    y=zeros(size(x))
    for i ∈ eachindex(x)
        y[i]=A*(x[i]^2)+B*x[i]+C #La ecuación en función del intervalo
        println(y[i]) 
    end
    plot(x,y,w=1)
        title!("Parabola")
        xaxis!("Eje x")
        yaxis!("Eje y")
end        
```

```julia
Parabola(1,0,0,1.0,-10.0,10.0)
```
