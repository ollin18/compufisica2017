```julia
function Recta(m,b,a,c,d) #m es la pendiente, b la ordenada al origen, a y d el valor inicial y final del intervalo, c el espacio entre cada numero del intervalo
    X = zeros(collect(a:c:d)) #Este es un arreglo de tamaño a-d con intervalos de c
    Y = zeros(collect(a:c:d)) 
    for i ∈ eachindex(collect(a:c:d)) #Esto hara que se hagan las siguientes operaciones en el arreglo
        X[i] = collect(a:c:d)[i] #Llamamos el arreglo de a-d 
        for i ∈ eachindex(collect(a:c:d)) 
            Y[i]=(m*collect(a:c:d)[i])+b #aplicamos la regla de correspondencia "y=mx+b"
        end
        recta = hcat(X,Y) #se concatenan los arreglos
        writedlm("recta.csv",recta,'|') #Aqui lera el archivo
        end
    print(X,Y) #E imprimimos las variables
end
```

```julia
function Parabola(a,b,c,n,m,l) #a,b,c son de; "y=ax^2+bx+c", m y l son los valores iniciales y finales del intevalo y e el espacio entre estos
    X = zeros(collect(n:m:l)) #Creamos un arreglo de n a l
    Y = zeros(collect(n:m:l))
    for i ∈ eachindex(collect(n:m:l)) #Esta operacion se realizara para cada elemento del arreglo
        X[i] = collect(n:m:l)[i] #Nombramos los arreglos
        for i ∈ eachindex(collect(n:m:l)) #aquí le pedimos que todas las operaciones se apliquen a cada elemento del arreglo
            Y[i]=(a*(X[i])^2)+b*X[i]+c #Y aplicamos "y=ax^2+bx+c"
        end
        parabola = hcat(X,Y) #Concatenamos los arreglos 
        writedlm("parabola.csv",parabola,'|') #Y se crea un archivo con los datos
        end
   print(X) #Y pedimos que imprima losarreglos
   print(Y)
end
```
