```julia
function Recta(m,b,a,c,d) #m hace referencia a la pendiente, b a la ordenada al origen, a y d el valor inicial y final del intervalo respectivamente, c establece el espacio entre cada numero del intervalo
    Y = zeros(collect(a:c:d)) #Aqui definimos un arreglo de ceros que tendra el tamaño de un areglo que ira desde "a" hasta "d", avanzando "c" espacios entre cada número
    X = zeros(collect(a:c:d)) #Aquí hacemos lo mismo que arriba 
    for i ∈ eachindex(collect(a:c:d)) #Aquí le decimos que aplique las siguientes operaciones para cada elemento en el arreglo que va de a hasta d
        X[i] = collect(a:c:d)[i] #Aquí igualamos el arreglo X al arreglo que va de a hasta d
        for i ∈ eachindex(collect(a:c:d)) #Aqui le decimos que aplique las siguientes operaciones a cada elemento del arreglo (a,d)
            Y[i]=(m*collect(a:c:d)[i])+b #Aquí aplicamos la regla de correspondencia de una recta que corresponde con la forma "y=mx+b"
        end
        recta = hcat(X,Y) #Aquí concatenamos ambos arreglos 
        writedlm("recta.csv",recta,'|') #Aquí escribimos la concatenación en un archivo para su lectura 
        end
    print(X,Y) #le pedimos que devuelva X y Y
end
```

```julia
function Parabola(a,b,c,d,e,f) #a,b,c son los coeficientes de la parabola con forma "y=ax^2+bx+c", d y f son los valores iniciales y finales del intevalo y "e" se refiere al espacio entre cada número del intervalo
    Y = zeros(collect(d:e:f)) #Aquí hacemos un arreglo de ceros del tamaño de otro arreglo con valores desde "d" hasta "f"
    X = zeros(collect(d:e:f)) #Aquí hacemos algo analogo a arriba
    for i ∈ eachindex(collect(d:e:f)) #Aquí le decimos que aplique todas las operaciones que siguen a cada elemento del arreglo
        X[i] = collect(d:e:f)[i] #aquí decimos que el arreglo de ceros X sea igual al arreglo que va desde d hasta f
        for i ∈ eachindex(collect(d:e:f)) #aquí le pedimos que todas las operaciones se apliquen a cada elemento del arreglo
            Y[i]=(a*(X[i])^2)+b*X[i]+c #Aquí aplicamos la regla de correspondencia de la parábola "y=ax^2+bx+c"
        end
        parabola = hcat(X,Y) #Aquí concatenamos los arreglos X Y despues de efectuar las operaciones anteriores
        writedlm("parabola.csv",parabola,'|') #Aquí escribimos un archivo con los datos arrojados
        end
   print(X,Y) #Aquí le pedimos nos regrese los arreglos
end
```
