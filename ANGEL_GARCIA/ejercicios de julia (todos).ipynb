```julia
function(factorialde(N)) #definimos la función 
    if N ≤ 1 # si  "N" es 0 o 1 el factorial corresponde a 1 
        1
    else
        N*factorialde(N-1) 
        #cuando es diferente la funcion se vuelve recursiva y regrasa hasta un numero que puede calcular yde ahi continua
    end
end
```

```julia
#ejemplo de cuando es 0  o  1
factorialde(1)
```

```julia
#ejemplo de cuando no 
factorialde(10)
```

```julia
function(aproxPi(N)) #definimos la función
    aprox=0
    for n ∈ 0:N
        aprox = aprox + (-1)^n/((2*n)+1) #la suma se actualiza con el valor del su¿iguiente sumando
    end
    aprPi = 4*aprox 
    println(aprPi)
end
```

```julia
function chicharonera(a,b,c) #definimos la funcion
   if b^2-4*a*c > 0 #identifica si el discriminante es mayor a cero y evalua la formula general 
        print("raiz real")# imprime que se trata de una raiz real
         X₋ =(-b-sqrt(b^2-4*a*c))/2*a#evalua el caso de la raiz negativa 
         X⁺ =(-b+sqrt(b^2-4*a*c))/2*a#evalua el caso de la raiz positiva
    X = hcat(X⁺,X₋)
    print(X)#por ultimo imprime ambos resultados
    
    end
    if b^2-4*a*c < 0#identifica si el discriminante es menor a cero y evalua la formula general multiplicando por un numero complejo
        print("raiz compleja")# imprime que se trata de una raiz compleja
         X₋ =(-b-sqrt((1im)*b^2-4*a*c))/2*a#evalua el caso de la raiz negativa
         X⁺ =(-b+sqrt((1im)*b^2-4*a*c))/2*a#evalua el caso de la raiz positiva
    X = hcat(X⁺,X₋)
    print(X)#por ultimo imprime ambos resultados
    end
    if b^2-4*a*c == 0#identifica si el discriminante es igual a cero y evalua la formula general 
        print("raiz doble")# imprime que se trata de una raiz doble
         X₋ =(-b-sqrt(b^2-4*a*c))/2*a#evalua el caso de la raiz negativa
         X⁺ =(-b+sqrt(b^2-4*a*c))/2*a#evalua el caso de la raiz positiva
    X = hcat(X⁺,X₋)
    print(X)#por ultimo imprime ambos resultados
    
    end
   
end
```

```julia
#ejemplo de raiz real
chicharonera(3,5,2)
```

```julia
#ejemplo de raiz compleja
chicharonera(6,5,2)
```

```julia
#ejemplo de riz doble
chicharonera(1,2,1)
```

```julia
function prodint(a,b)
    producint = 0
if length(a) == length(b)
    for i ∈ eachindex(a,b)
    producint =producint + a[i]*b[i] 
    end
    println(producint)

else
    "error tamaño de los vectores distinto"
end
end
```

```julia
function MRU(v,t) #definimos la funcion en terminos de la velocidad y el tiempo
posicion = zeros(length(t))#el tamaño debe ser el mismo que el del tiempo
for i ∈ 1:length(t)
    if i != 1 #para todas las posiciones excepto la primera se actualizarán atravez de laformula de la velocidad
        posicion[i] = posicion[i-1]+v*Δt
    end
end
    MRU=hcat(t,posicion) #se cea un arreglo de la posicion contra el tiempo
writedlm("datosMRU.txt",MRU,'|') #se escribe el archivo
readdlm("datosMRU.txt",'|') #lo lee
end
    
```

```julia
#ejemplo
Δt = 0.25
t = 0:Δt:30
MRU(58,t);
```

```julia
function caidalibre(t,y0) #definimos la funcion enterminos del tiempo y la posicion inicial
    g=9.81 #definimos las constantes
    Y=zeros(t)#el tamaño del arreglo para Y
    for i ∈ eachindex(t)#realiza la operaciones convenientes para el movimiento visto desde la perspectiva vertical (Y)
        Y[i]=y0-((g*t[i]^2)/2)
    end
 caidlibr=hcat(t,Y) #se cea un arreglo de la posicion en Y contra el tiempo
writedlm("datoscaidalibre.txt",caidlibr,'|') #se escribe el archivo
readdlm("datoscaidalibre.txt",'|') #lo lee
end    
```

```julia
#ejemplo
t = 0:.25:40
y0=200
caidalibre(t,y0)
```

```julia
function Tiroparabolico(t,θ,v0)#definimos la funcion enterminos del tiempo el angulo y la velocidad inicial
g = 9.81                  #definimos las constantes
v = v0*cos(θ)+v0*sin(θ) #la velocidad como vector
X = zeros(t)#el tamaño del arreglo para X
Y = zeros(t)#el tamaño del arreglo para Y
for i ∈ eachindex(t) #realiza la operaciones convenientes para el movimiento visto desde la perspectiva horizontal (X) y la perspectiva vertical (Y)
    X[i]=v0*cos(θ)*t[i]
    Y[i]=(-1(0.5*g*t[i]^2)+v0*sin(θ)*t[i])
end
     tiro_parabol=hcat(t,X,Y) #se cea un arreglo de las posiciones en X y Y contra el tiempo
writedlm("datostir_parabol.txt",tiro_parabol,'|')#se escribe el archivo
readdlm("datostir_parabol.txt",'|')#lo lee
    
end
```

```julia
#ejemplo
t = 0.0:0.25:100
θ=π/16
v0=36
Tiroparabolico(t,θ,v0)
```
