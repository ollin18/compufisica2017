**a)** Escribe un programa para calcular una recta de valores de Y en términos de valores X en cierto rango de X, i.e. el programa debe arrojar como resultado un dato para graficar los puntos (x,y) de una recta. 

El usuario debe introducir los para metros de la recta (pendiente m y ordenada la origen b), así como los valores inicial, final y de tamaño del intervalo para x.

--------------------------------

Continuando con el tema de la recta, hay muchas formas de trabajar con una recta en el plano, en este caso nosotros utilizaremos la formula de la pendiente y ordenada al origen.
$$ y=m(x)+b$$

Donde:

$y(t)$ es una variable dependiente de $x$

$x(t)$ es la variable independiente

$m$ es la pendiente de la recta que esta definida como $m=\frac{y_2 - y_1}{x_2 - x_1}$

$b$ es la abscisa cuando se tiene la coodenada $(0,b)\in R$


```julia
#Primero definimos nuestro TiempoR que es con respecto a lo que avanzara x(t)

t1=0:0.125:30
collect(t1);
```

```julia
#Definimos sobre que van a trabajar; x(t) con respecto al Dominio y(t).

Dom1=zeros(t1);
```

```julia
#Damos valores arbitrarios para la pendiente y la coordenada de la ordenada al origen

m=3; b=14;
```

```julia
#Decimos que para cada elemento del tiempoR tenemos que x=t*m + b; y que y=t

for i ∈ eachindex(t1)
    Dom1[i]=t1[i]*m+b   
end
```

```julia
using Plots;
gr()
```

```julia
plot(t1,Dom1,w=2,label=["Recta"])
title!("Recta")
xaxis!("X")
yaxis!("Y")
```

```julia
#Definimos nuestra funcion recta en funcion de la pendiente, la ordenada al origen y el Dominio t1.

function Recta(m,b,t1)
    Dom1=zeros(t1)
    for i ∈ eachindex(t1)
        Dom1[i]=t1[i]*m+b   
    end
    Dom1
end
```

```julia
R1=Recta(9,4,t1)
R2=Recta(4,23,t1)
R3=Recta(-1,213,t1)
R=hcat(R1,R2,R3)
```

```julia
plot(t1,R,w=2)
title!("Rectas")
xaxis!("X")
yaxis!("Y")
```

**b)** Escribe un programa para calcular una parábola de valores Y en términos de valores X en cierto rango de X, i.e. el programa debe arrojar como resultado un archivo con los datos para graficar los puntos (x,y) de una parábola. 

El usuario debe introducir los parámetros de la ecuación cuadrática, así como los valores inicial, final y de tamaño del intervalo para X.

Ahora con la parábola, utilizaremos la formula de la parábola en su forma canónica. (suponiendo que el vertice esta en el origen)

$$ y^2=4px$$

Donde:

$y$ es una variable dependiente de $x$

$x$ es la variable independiente

$p$ es la distancia de el vertice al foco o del vertice al origen

Después de unos despejes tenemos que:

$$ y=\pm2\sqrt{px}$$

```julia
#Como la funcion tiene a x como la variable indepentiente, crearemos dos vectores del mismo tamaño llamados Y1 (para las positivas) y Y2 (para los negativos)

function Parabola(x,p)
    Y1=zeros(x)
    Y2=zeros(x)
    for i in eachindex(x)
        Y1[i]=+2*((p*x[i])^(1/2))
        Y2[i]=-2*((p*x[i])^(1/2))
    end
    T=hcat(x,Y1,Y2)
    writedlm("Parabola.csv",T,'|')
    readdlm("Parabola.csv",'|')
end
```

```julia
#Definimos nuestro rango de X

X=collect(0:0.125:50)
```

```julia
#Damos a P como la matriz solucion

P=Parabola(X,4)
```

```julia
using Plots;
gr()
```

```julia
#Ploteamos y listo!!

plot(P[1:end,2:3],P[1:end,1],w=3)
title!("Parábola")
xaxis!("Y")
yaxis!("X")
```
